```python
import pandas as pd
import openpyxl as xl

df = pd.read_excel("data/cryptomarkets_dataset.xlsx", sheet_name="data", parse_dates=["date"], index_col="date")
# df = df.asfreq("D")
df = df.sort_index()

# Dropping unwanted columns
df = df.drop(df.columns[0], axis=1)
df = df.drop(['Time'], axis = 1)
# Unparsed data
object_col_name = df.select_dtypes(include=['object']).columns
df[object_col_name] = df[object_col_name].apply(lambda x: pd.to_numeric(x, errors='coerce'))

df.head()
```

```python
# Export data in csv
df.to_csv("data/output_data.csv", index=True, header=True)
```
