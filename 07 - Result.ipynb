```python
# Import all data visualisation libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.patches as mpatches
import matplotlib.dates as mdates
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import seaborn as sns
import ipywidgets as ipy
from IPython.display import Image

sns.set_theme()
sns.set_palette("colorblind")
sns.set_style('whitegrid')
plt.tight_layout()
%matplotlib inline
```

```python
df = pd.read_csv('data/pred_results.csv')
df.set_index('date', inplace=True)
df.index = pd.to_datetime(df.index)
```

```python
targets = ['BTC', 'ETH', 'DOGE', 'SHIB', 'DOT', 'SOL', 'ADA', 'MATIC', 'BNB', 'LTC', 'XRP']
df.columns
```

```python
for target in targets:
    df[f'_{target}_real_rend'] = np.log(df[f'_{target}_real']).diff()
    df[f'_{target}_pred_rend'] = np.log(df[f'_{target}_pred']).diff()
    
df
```

```python
for i in targets:
    sub_df = df[[f'_{i}_real_rend', f'_{i}_pred_rend', f'_{i}_pred', f'_{i}_real']].dropna()
    # Compute RMSE between real and pred rend


    rmse = np.sqrt(np.mean((sub_df[f'_{i}_real_rend'] - sub_df[f'_{i}_pred_rend'])**2))
    print(f'RMSE between real and pred rend for {i} is {rmse}')
    
    # Plot pred and real rend
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(sub_df[f'_{i}_real_rend'], label='ObservÃ© rendement')
    ax.plot(sub_df[f'_{i}_pred_rend'], label='Pred rendement')
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(mdates.AutoDateLocator()))
    # Add legend
    ax.legend()
    # Add title and labels
    ax.set_title(f'Ã‰volution du rendement du {i}')
    
    # Plot pred and real value
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(sub_df[f'_{i}_real'], label='ObservÃ© valeur')
    ax.plot(sub_df[f'_{i}_pred'], label='Pred valeur')
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(mdates.AutoDateLocator()))
    # Add legend
    ax.legend()
    # Add title and labels
    ax.set_title(f'Ã‰volution de la valeur du {i}')
```

```python
# Display in table RMSE of pred_rend and real_rend for each crypto

# Compute RMSE between real and pred rend
rmse = []
for i in targets:
    sub_df = df[[f'_{i}_real_rend', f'_{i}_pred_rend', f'_{i}_pred', f'_{i}_real']].dropna()
    rmse.append(np.sqrt(np.mean((sub_df[f'_{i}_real_rend'] - sub_df[f'_{i}_pred_rend'])**2)))
    
# Make rmse[:-3] and targets[:-3] in a dataframe
result = pd.DataFrame({'Crypto': targets[:-3], 'RMSE': rmse[:-3]}).sort_values(by='RMSE', ascending=False)
# Sort result by index value
result = result.sort_index()

result
```
